```
'''
	Brayan Alejandro Duarte Acevedo - PeopleSoftID
	Program #1: Login System
	COSC 1306, Fall 2017 '''

'''He marcado mis notas con un <A> al final para ser distinguidas más fácilmente por cualquier lector incluyendome.
   I've marked my notes with an <A> at the endline to be more easily distinguished by any reader including me.'''
print("----------------------------------------------")
print("\n\tCreate Account\n")
username = input("Please enter a new username: ")
password = input("Please enter a new password: ")
print("\nAccount created!\n")
run = True

while run is True:  # while run is true, program will keep asking for sign in
	login_success = False
	login_check = input("Would you like to log in? (y/n): ")
	if login_check.lower() == 'y':  # .lower() converts input to lowercase
		print("You will have 3 attempts.")
    
		for attempt_num in range(3,0,-1):  # I've used a range without var name cause I've considered it is better (each for loop to login have their own attempts range) <A>

			print("\n----------------------------------------------\n")
			print("\tLogin\n")
			check_user = input("Username: ")
			check_pw = input("Password: ")

			# if login information is correct
			if(check_user == username and check_pw == password):
				print("\nLogin succeeded!")
				print("\n----------------------------------------------\n")
				login_success = True
				break  # this break stmt, if reached, program will exit the for loop

			else:  # if login information is incorrect
				print("Incorrect credentials.")

				tries_left = attempt_num -1

				print('\nYou have', tries_left, 'attempts remaining.')

				if tries_left == 0:
					run = False
					break

		while login_success is True:		# if login is successful, go to menu
			print("\tMenu\n")
			print("1. Change Username")
			print("2. Change Password")
			print("\nPress any other key to log out.")
			choice = input("\nEnter number of a menu option: ")
			print("\n----------------------------------------------\n")

			if choice == "1":  # if user selects Change Username
				print("\tChange Username\n")
				check_success = False

				for attempt_num in range(3,0,-1):
					check_user = input("Enter your current username: ")
					check_pw = input("Enter your current password: ")
					# if check information is correct <A>
					if(check_user == username and check_pw == password):
						check_success = True
						break  # this break stmt, if reached, program will exit the for loop

					else:  # if check information is incorrect <A>
						print("\nIncorrect credentials.")

						tries_left = attempt_num -1

						print('You have', tries_left, 'attempts remaining.\n')

						if tries_left == 0:
							login_success = False
							break
          
				if check_success == True:
					new_username = input("Enter a new username: ")
					new_username_check = input("please confirm new username: ")
					while new_username != new_username_check: #if new username is different to new username check
						print('\nnew user confirmation is wrong, please try again:\n')
						new_username = input("Enter a new username: ")
						new_username_check = input("please confirm new username: ")
					username = new_username
					print("\nYour name is now: ",username,"\n")
				print("\n----------------------------------------------\n")

			elif choice == "2":  # if user selects Change Password
				print("\tChange Password\n")
				check_success = False

				for attempt_num in range(3,0,-1):
					check_user = input("Enter your current username: ")
					check_pw = input("Enter your current password: ")
					# if check information is correct <A>
					if(check_user == username and check_pw == password):
						check_success = True
						break  # this break stmt, if reached, program will exit the for loop

					else:  # if check information is incorrect <A>
						print("\nIncorrect credentials.")

						tries_left = attempt_num -1

						print('You have', tries_left, 'attempts remaining.\n')

						if tries_left == 0:
							login_success = False
							break
          
				if check_success == True:
					new_password = input("Enter a new password: ")
					new_password_check = input("please confirm new password: ")
					while new_password != new_password_check: #if new password is different to new password check
						print('\nnew password confirmation is wrong, please try again:\n')
						new_password = input("Enter a new password: ")
						new_password_check = input("please confirm new password: ")
					password = new_password
					print("\nYour password has been changed. Returning to main menu.\n")
        
				print("\n----------------------------------------------\n")

			else:  # This section is reached if user does not enter 1 or 2 at menu
				login_success = False
				print("Logging out...")
				print("\n----------------------------------------------\n")

	else:  # this is reached when the user does not enter y or Y at login prompt
		run = False

# This line is only reached after the user chooses to exit the program
# This is outisde of the main loop
input("Program exiting, press any key to continue...")
quit() 

```
